# 4. График функции одной переменной
## Необходимо написать программу построения графика функции одной переменной

```python
%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.axes_grid.axislines import SubplotZero

def f(m, b):
    plt.figure(2)
    x = np.linspace(-15, 10, num=1000)
    plt.plot(x, m * x + b)
    plt.ylim(-15, 15)
    plt.show()



def g(n):
    fig = plt.figure(figsize=(20,10))
    ax = SubplotZero(fig, 111)
    fig.add_subplot(ax)

    for direction in ["xzero", "yzero"]:
        ax.axis[direction].set_axisline_style("-|>")
        ax.axis[direction].set_visible(True)

    for direction in ["left", "right", "bottom", "top"]:
        ax.axis[direction].set_visible(False)
    
    x = np.linspace(-5, 5., 100)
#     print(x)
    ax.plot(x, (x+n)*(x-(n+1))*(x-(n+2)))
    plt.ylim(-15, 15)
    plt.xlim(-3, 5)
    plt.show()


    
# interactive_plot = interactive(f, m=(-2.0, 2.0), b=(-3, 3, 0.5))
interactive_plot = interactive(g, n=(-2.0, 2.0))
output = interactive_plot.children[-1]
output.layout.height = '350px'
interactive_plot
```

# 5. График функции двух переменных
## Необходимо написать программу построения графика функции двух переменных

![image](5_condition.png)

```python
import math
def two_var_plot():
    fig = plt.figure(figsize=(20,10))
    ax = SubplotZero(fig, 111)
    fig.add_subplot(ax)

    for direction in ["xzero", "yzero"]:
        ax.axis[direction].set_axisline_style("-|>")
        ax.axis[direction].set_visible(True)

    for direction in ["left", "right", "bottom", "top"]:
        ax.axis[direction].set_visible(False)
    
    x = np.linspace(0, 2*np.pi, 400)
    y = np.linspace(0, 2*np.pi, 400)
    alpha = (x-np.pi)**2+(y-np.pi)**2
    z = (1/5)*np.sin(x)*np.cos(y)-(3/2)*np.cos(7*alpha[0]/4)*math.exp(-alpha[0])
#     print(z)
    ax.plot(x, z)
    plt.ylim(-0.5, 0.5)
    plt.xlim(0, 5)
    plt.show()

two_var_plot()
```
